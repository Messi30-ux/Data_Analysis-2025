```python
#Data Aggregation
Price = [100000.56, 180000.57, 3000.59, 500000.66, 5000.18]
sum_fx = sum(Price)
print(sum_fx)
```

```python
fx = len(Price)
print (fx)
```

```python
mean_Price = sum_fx / fx
```
